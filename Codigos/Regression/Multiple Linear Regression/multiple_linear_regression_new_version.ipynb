# Eliminación hacia atrás utilizando solamente p-valores:

```python
import statsmodels.api as sm
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('50_Startups.csv')
dataset
```

```python
X = pd.get_dummies(dataset,columns=["State"], drop_first=True,prefix=["0"]).drop(columns=["Profit"])
display(X.head())
X = pd.DataFrame(X, columns=sorted(X.columns))
display(X.head())
X = X.to_numpy()
display(X[:5,:])
X = np.append(arr = np.ones((50,1)).astype(int), values = X, axis = 1)
X[:5,:]
```

```python
y=dataset["Profit"]
y[:5]
```

```python
def backwardElimination(x, sl):    
    numVars = len(x[0])    
    for i in range(0, numVars):        
        regressor_OLS = sm.OLS(y, x.tolist()).fit()        
        maxVar = max(regressor_OLS.pvalues) #.astype(float)        
        if maxVar > sl:            
            for j in range(0, numVars - i):                
                if (regressor_OLS.pvalues[j] == maxVar):     #.astype(float) == maxVar):                    
                    x = np.delete(x, j, 1)    
    print(regressor_OLS.summary())    
    return x 
```

El código proporcionado implementa un algoritmo de eliminación hacia atrás para seleccionar las variables más significativas en un modelo de regresión lineal. 

El algoritmo toma dos parámetros: "x", que es una matriz de características independientes, y "sl", que es el nivel de significancia (p-value) máximo permitido. El objetivo del algoritmo es eliminar las variables que tienen un p-value mayor que el nivel de significancia dado.

El código comienza obteniendo el número de variables en la matriz "x" y luego inicia un bucle que itera sobre todas las variables. Dentro del bucle, se ajusta un modelo de regresión lineal utilizando la función OLS de la biblioteca "sm" y se guarda en la variable "regressor_OLS".

A continuación, se determina la variable con el p-value máximo utilizando la función "max" en la lista de p-values del modelo ajustado. Si este valor máximo es mayor que el nivel de significancia dado, se inicia otro bucle que itera sobre las variables restantes.

Dentro de este segundo bucle, se verifica si el p-value de cada variable es igual al valor máximo encontrado anteriormente. Si es así, se elimina esa variable de la matriz "x" utilizando la función "np.delete".

Después de que el bucle principal se completa, se imprime el resumen del modelo ajustado final utilizando la función "summary" de la variable "regressor_OLS". Finalmente, se devuelve la matriz "x" modificada, que ahora contiene solo las variables más significativas según el criterio de eliminación hacia atrás.

En resumen, el código implementa un algoritmo de eliminación hacia atrás para seleccionar las variables más significativas en un modelo de regresión lineal y devuelve la matriz de características independientes modificada.

### que es p.value?

El p-value (valor p) es una medida estadística utilizada en pruebas de hipótesis para evaluar la evidencia en contra de una hipótesis nula. En el contexto de la regresión lineal, el p-value se utiliza para determinar la significancia estadística de los coeficientes de las variables independientes.

En términos generales, el p-value indica la probabilidad de obtener un resultado igual o más extremo que el observado, asumiendo que la hipótesis nula es cierta. Si el p-value es menor que un nivel de significancia predefinido (generalmente 0.05 o 0.01), se considera que hay suficiente evidencia para rechazar la hipótesis nula y concluir que el coeficiente de la variable independiente es significativamente diferente de cero.

En el contexto de la eliminación hacia atrás en el código que proporcionaste, se utiliza el p-value para determinar qué variables deben eliminarse del modelo de regresión lineal. Si el p-value de una variable es mayor que el nivel de significancia dado (sl), se considera que esa variable no es estadísticamente significativa y se elimina del modelo.

```python
SL = 0.05
X_opt = X[:, [0, 1, 2, 3, 4, 5]]
X_Modeled = backwardElimination(X_opt, SL)
```

# Eliminación hacia atrás utilizando  p-valores y el valor de  R Cuadrado Ajustado:

## que es el R Cuadrado Ajustaso:

El valor de R cuadrado ajustado (R-squared ajustado) es una medida utilizada para evaluar la calidad de ajuste de un modelo de regresión. Es similar al valor de R cuadrado (R-squared), pero tiene en cuenta el número de variables independientes en el modelo y penaliza el valor de R cuadrado por la inclusión de variables irrelevantes.

El R cuadrado ajustado se calcula utilizando la siguiente fórmula:

R-cuadrado ajustado = 1 - [(1 - R-cuadrado) * (n - 1) / (n - k - 1)]

Donde:
- R-cuadrado es el valor de R cuadrado del modelo (que indica la proporción de la variabilidad de la variable dependiente explicada por las variables independientes).
- n es el número de observaciones en el conjunto de datos.
- k es el número de variables independientes en el modelo.

El R cuadrado ajustado toma valores entre 0 y 1. Un valor más cercano a 1 indica un mejor ajuste del modelo, teniendo en cuenta el número de variables independientes. Se prefiere un R cuadrado ajustado más alto, ya que indica que una mayor proporción de la variabilidad de la variable dependiente es explicada por las variables independientes relevantes en el modelo.

```python
def backwardElimination(x, SL):    
    numVars = len(x[0])    
    temp    = np.zeros((50,6)).astype(int)    
    for i in range(0, numVars):        
        regressor_OLS = sm.OLS(y, x.tolist()).fit()        
        maxVar        = max(regressor_OLS.pvalues)      #.astype(float)        
        adjR_before   = regressor_OLS.rsquared_adj      # .astype(float)        
        if maxVar > SL:            
            for j in range(0, numVars - i):                
                if (regressor_OLS.pvalues[j] == maxVar): #astype(float)                    
                    temp[:,j]     = x[:, j]                    
                    x             = np.delete(x, j, 1)                    
                    tmp_regressor = sm.OLS(y, x.tolist()).fit()                    
                    adjR_after    = tmp_regressor.rsquared_adj  # .astype(float)                    
                    if (adjR_before >= adjR_after):                        
                        x_rollback = np.hstack((x, temp[:,[0,j]]))                        
                        x_rollback = np.delete(x_rollback, j, 1)     
                        print (regressor_OLS.summary())                        
                        return x_rollback                    
                    else:                        
                        continue    
    #print(regressor_OLS.summary())    
    return x 
```
